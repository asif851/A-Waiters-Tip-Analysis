```python
#import modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
```

```python
#load dataset

df=sns.load_dataset('tips')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
lowest_tip=df['tip'].min()
higest_tip=df['tip'].max()
avr_tip=df['tip'].mean()

print(f"Highest Tip :{higest_tip}\nLowest Tip  :{lowest_tip}\nAverage Tip :{avr_tip}")
```

```python
bill_sum=df['total_bill'].sum()
tip_sum=df['tip'].sum()
percent=tip_sum/bill_sum*100

print(f"Average tip is {percent}% of total bill.")
```

```python
corrMatrix=df.corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```
